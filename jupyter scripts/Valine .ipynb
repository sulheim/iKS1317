```python
import cobra
# Load model
model_path = "P://git/gem_sco/modelS.xml"
model = cobra.io.read_sbml_model(model_path)
```

```python
reaction_ids = ["VALDxi", "VALTRS", "VALTA"]
vdh_mutant = model.copy()
vdh_mutant.genes.SCO4089.knock_out()
vdh_mutant.genes.SCO4089.reactions
vdh_mutant.reactions.EX_nh4_e.bounds = (0, 1000)
vdh_mutant.reactions.EX_val__L_e.bounds = (-1,1000)

vdh_mutant.optimize()
```

```python
for r_id in reaction_ids:
    r = model.reactions.get_by_id(r_id)
    print(r.gene_reaction_rule)
```
